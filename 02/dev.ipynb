```python
def get_data(frac: float = .5) -> pd.DataFrame:
    dataset = pd.read_parquet(DATASET_PATH)
    df = dataset.sample(frac=frac)
    del dataset

    df["PU_DO"] =  df["PULocationID"].astype(str) + "_" + df["DOLocationID"].astype(str)
    df['duration'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60
    cols = df.columns.tolist()
    df = pd.get_dummies(df, columns=['PU_DO'])
    ohed_cols = [col for col in df.columns if col not in cols]

    X = df[ohed_cols + ['trip_distance']] 
    y = df['duration']

    return X, y
```

```python
X, y = get_data(frac=0.5)
```

```python
import mlflow


mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)
mlflow.set_experiment("yellow-taxi-trip-duration")
```

```python
experiment_name = "yellow-taxi-trip-duration"
exps = mlflow.search_experiments(filter_string=f"name = '{experiment_name}'")
if len(exps) > 0:
    mlflow.set_experiment(experiment_name)
else:
    mlflow.create_experiment(experiment_name)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import root_mean_squared_error, r2_score


with mlflow.start_run():
    mlflow.set_tag("developer", "Alper")
    mlflow.set_tag("model", "Lasso")
    
    # parameters for the experiment
    dataset_frac = 0.5
    test_size = 0.2
    alpha = 0.1

    # log these parameters
    mlflow.log_param("dataset_frac", 0.5)
    mlflow.log_param("test_size", 0.2)
    mlflow.log_param("alpha", 0.1)
    
    # dataset preparation
    X, y, numerical_features = get_data(frac=dataset_frac)
    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    mlflow.log_param("additional_features", numerical_features)
    # training the model
    model = Lasso(alpha=alpha)
    model.fit(x_train, y_train)

    # get the preds
    train_preds = model.predict(x_train)
    valid_preds = model.predict(x_test)

    # calculate the metrics
    training_rmse = root_mean_squared_error(y_train, train_preds)
    validation_rmse = root_mean_squared_error(y_test, valid_preds)

    training_r2 = r2_score(y_train, train_preds)
    validation_r2 = r2_score(y_test, valid_preds)

    # log the metrics
    mlflow.log_metric("training_rmse", training_rmse)
    mlflow.log_metric("validation_rmse", validation_rmse)
    mlflow.log_metric("training_r2", training_r2)
    mlflow.log_metric("validation_r2", validation_r2)
    
    print(f"Training RMSE: {training_rmse}")
    print(f"Validation RMSE: {validation_rmse}")

    print(f"Training R2: {training_r2}")
    print(f"Validation R2: {validation_r2}")
```
